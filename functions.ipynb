# Funktionen / *Functions*
Python-Funktionen sind ein Block von Anweisungen, der eine bestimmte Aufgabe ausführt. Die Idee besteht darin, häufig oder wiederholt durchgeführte Aufgaben zusammenzufassen und eine Funktion zu erstellen, sodass wir anstelle des wiederholten Schreibens desselben Codes für unterschiedliche Eingaben die Funktion aufrufen können, um den darin enthaltenen Code immer wieder zu verwenden.
Die folgenden Schritte beschreiben, wie man eine Funktion definiert:

- Wir verwenden das Schlüsselwort def, um die Funktion zu deklarieren, gefolgt vom Funktionsnamen.
- Wir fügen Parameter zur Funktion hinzu: sie sollten innerhalb der Klammern der Funktion stehen. Die Zeile wird mit einem Doppelpunkt beendet.
- Die erste Anweisung einer Funktion kann optional sein: die Dokumentationszeichenkette (docstring) der Funktion.
- Wir fügen die Anweisungen hinzu, die die Funktion ausführen soll.
- Wir beenden die Funktion mit einer return-Anweisung, wenn die Funktion etwas ausgeben soll. Ohne die return-Anweisung gibt die Funktion None zurück.

---

*Python Functions are a block of statements that does a specific task. The idea is to put some commonly or repeatedly done task together and make a function so that instead of writing the same code again and again for different inputs, we can do the function calls to reuse code contained in it over and over again.*

*The following are the steps to define a function:*
- *Use the keyword def to declare the function and follow this up with the function name.*
- *Add parameters to the function: they should be within the parentheses of the function. End your line with a colon.*
- *The first statement of a function can be an optional statement; the documentation string of the function or docstring.*
- *Add statements that the functions should execute.*
- *End your function with a return statement if the function should output something. Without the return statement, your function will return `None`.*

Syntax:
```
def function_name(parameters):
    """Docstring"""
    # body of the function
    # optional return statement
    return expression
```

````python
# a function may not have parameters
# can be as simple as printing a statement
def fun():
    print("Welcome to fun with python")
````

````python
fun() # Driver code to call a function
````

# Docstring

Docstrings beschreiben, was eine Funktion macht, zum Beispiel den erwarteten Datentyp der Eingaben, die durchgeführten Berechnungen oder die Rückgabewerte. Diese Beschreibungen dienen als Dokumentation der Funktion, sodass jede Person, die den Docstring der Funktion liest, versteht, was die Funktion macht, ohne den gesamten Code in der Funktionsdefinition nachverfolgen zu müssen.

Funktions-Docstrings werden direkt in der Zeile nach dem Funktionskopf eingefügt und stehen zwischen dreifachen Anführungszeichen.

Um den Docstring einer Funktion in Python zu überprüfen, kann die eingebaute Funktion `help()` verwenden werden oder auf das Attribut `__doc__` der Funktion zugegriffen werden.

---

*Docstrings describe what your function does, such as the data type of the input it expects, computations it performs or its return values. These descriptions serve as documentation for your function so that anyone who reads your function’s docstring understands what your function does, without having to trace through all the code in the function definition.*

*Function docstrings are placed in the immediate line after the function header and are placed in between triple quotation marks.*

*To check the docstring of a function in Python, you can use the built-in `help()` function or access the function’s `__doc__` attribute.*


````python
def evenOdd(x):
    """
    This function calculates if a number is even or odd.
    input:
        x: int
    output:
        "Even" or "Odd": str
    """
    if (x % 2 == 0):
        return "Even"
    else:
        return "Odd"
````

# Parameter und Argumente / *Parameters and Arguments*

Beim Definieren einer Funktion wird innerhalb der Klammern eine Liste von Variablen angegeben, die als formale Parameter bezeichnet werden. Diese Parameter fungieren als Platzhalter für die Daten oder Argumente, die der Funktion beim Aufruf übergeben werden. Beim Aufrufen der Funktion muss jedem formalen Parameter ein Wert zugewiesen werden.

Je nachdem, wie die Argumente beim Definieren einer Python-Funktion angegeben werden, werden sie in die folgenden Kategorien eingeteilt:
* Positions- oder Pflichtargumente
* Schlüsselwortargumente
* Standardargumente
* Beliebige oder variabel lange Argumente

---

*While defining a function, you specify a list of variables (known as formal parameters) within the parentheses. These parameters act as placeholders for the data or arguments that will be passed to the function when it is called. When the function is called, value to each of the formal parameters must be provided.*

*Based on how the arguments are declared while defining a Python function, they are classified into the following categories:*
- *Positional or Required Arguments*
- *Keyword Arguments*
- *Default Arguments*
- *Arbitrary or Variable-length Arguments*

Bei **Positionsargumenten** werden die Werte den Parametern entsprechend ihrer Reihenfolge im Funktionsaufruf zugewiesen.

---

In **positional arguments**, values are assigned to parameters based on their order in the function call.

````python
def calculateProfit(item_name, cost, price):
    """
    calculates profit of an item.
    input:
        item_name: str
        cost: float|int
        price: float|int
    """
    profit = price - cost
    print(f'By selling {item_name}, you made a profit of {profit:.2f} euros.')
````

# Übergabe durch Referenz und Übergabe durch Wert / *Pass by Reference and Pass by Value*

In Python sind Variablen Referenzen auf Objekte. Wenn wir sie an eine Funktion übergeben, hängt das Verhalten davon ab, ob das Objekt veränderbar (mutable) oder unveränderbar (immutable) ist:
- Veränderbare Objekte (z. B. Listen, Wörterbücher): Änderungen innerhalb der Funktion wirken sich auf das ursprüngliche Objekt aus.
- Unveränderbare Objekte (z. B. Integer, Strings, Tupel): Der ursprüngliche Wert bleibt unverändert.

---

*In Python, variables are references to objects. When we pass them to a function, the behavior depends on whether the object is mutable (like lists, dictionaries) or immutable (like integers, strings, tuples).*
- *Mutable objects: Changes inside the function affect the original object.*
- *Immutable objects: The original value remains unchanged.*

````python
# Function modifies the third element of list
def double_the_item(x):
    x[2] *= 2

lst = [10, 11, 12, 13]
double_the_item(lst)
print(lst)   # list is modified
````

````python
# Function tries to modify an integer
def double_the_item(x):
    x = x * 2
    return x

a = 10
double_the_item(a)
print(a)     # integer is not modified
````

| Type       | Mutable? | Changes outside function? |
| ---------- | -------- | ------------------------- |
| int        | ❌        | No                        |
| float      | ❌        | No                        |
| string     | ❌        | No                        |
| list       | ✅        | Yes                       |
| dictionary | ✅        | Yes                       |

# *args

`*args` ermöglicht einer Funktion, eine beliebige Anzahl von Positionsargumenten zu akzeptieren, die als Tupel gespeichert werden.

---

*`*args` allows a function to accept any number of positional arguments, which are stored as a tuple.*

````python
def add_numbers(*args):
    total = 0
    for n in args:
        total += n
    return total

print(add_numbers(2, 4, 6))
print(add_numbers(10, 19, 6, 90, 100))
````

### **kwargs

`**kwargs` ermöglicht einer Funktion, eine beliebige Anzahl von Schlüsselwortargumenten zu akzeptieren, die als Wörterbuch gespeichert werden.

---

*`**kwargs` allows a function to accept any number of keyword arguments, which are stored as a dictionary.*

````python
def print_info(**kwargs):
    for key in kwargs:
        print(key, ":", kwargs[key])

print_info(name="Anna", age=25, city="Berlin")
print_info(name="Laptop", price=1200, brand="Dell")
````

Eine Funktion kann innerhalb ihres Rumpfes eine andere Funktion aufrufen, um Code wiederzuverwenden oder Operationen zu kombinieren.

---

*A function can call another function inside its body to reuse code or combine operations.*

````python
# function calling another function
def square(x):
    return x * x

def sum_of_squares(a, b):
    return square(a) + square(b)

print(sum_of_squares(2, 3))
````

---

*Moreover, in Python, functions are first-class objects, meaning they can be passed as arguments to other functions. Sometimes you want to apply different operations to the same data and passing a function as an argument allows a program to change its behavior dynamically by applying different operations without rewriting the main function.*

# Rekursive Funktionen / *Recursive Functions*

Eine rekursive Funktion ist eine Funktion, die sich selbst aufruft, um ein Problem zu lösen. Sie wird häufig in mathematischen Problemen oder bei Divide-and-Conquer-Ansätzen verwendet. Es ist immer wichtig, einen Basisfall einzuschließen, um unendliche Rekursion zu vermeiden.

Zwei wichtige Teile existieren immer:
* Basisfall (Base case) → der Punkt, an dem die Rekursion stoppt
* Rekursiver Fall (Recursive case) → der Punkt, an dem die Funktion sich selbst aufruft
  
---

*A recursive function is a function that calls itself to solve a problem. It is commonly used in mathematical and divide-and-conquer problems. Always include a base case to avoid infinite recursion.*

*Two important parts always exist:*
- *Base case → when the recursion stops*
- *Recursive case → when the function calls itself*

Beispiel: Berechne die Summe der n ersten natürlichen Zahlen:  $sum\_of\_numbers = 1 + 2+ 3 + ... + n$.

---

*Example: Compute sum of first n natural numbers: $sum\_of\_numbers = 1 + 2+ 3 + ... + n$.*

````python
def sum_numbers(n):
    # base case: we stop recursion
    if n == 1:
        return 1
    # recursive function call
    return n + sum_numbers(n-1)
````

````python
print(sum_numbers(4))
````

# Lambda Function

Lambda-Funktionen sind kleine anonyme Funktionen, das heißt, sie haben keinen definierten Namen. Sie sind kurze, kurzlebige Funktionen, die verwendet werden, um einfache Logik an eine andere Funktion zu übergeben.
- Sie enthalten nur einen Ausdruck.
- Das Ergebnis dieses Ausdrucks wird automatisch zurückgegeben (kein return-Schlüsselwort erforderlich).
  
---

*Lambda functions are small anonymous functions, meaning they do not have a defined name. These are small, short-lived functions used to pass simple logic to another function.*
- *Contain only one expression.*
- *Result of that expression is returned automatically (no return keyword needed).*

Syntax:
```
lambda arguments: expression
```

Examples:

```
square = lambda x: x * x
print(square(5))

# Output: 25
```

```
add = lambda a, b: a + b
print(add(3, 4))

# Output: 7
```
